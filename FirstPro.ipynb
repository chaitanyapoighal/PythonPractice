```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\inspiron\Desktop\Road Accident Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.columns
```

```python
acc_days=sns.countplot(x= 'Day_of_Week' ,data=df)
for bars in acc_days.containers:
     acc_days.bar_label(bars)
```

```python
We can see that on fridays the No.of accidents are more and next is Tuesday
```

```python
acc_area=sns.countplot(x= 'Urban_or_Rural_Area' ,data=df)
sns.set(rc={'figure.figsize':(15,5)})
for bars in acc_area.containers:
     acc_area.bar_label(bars)
```

```python
With above bar graph we can understand that Compare to Rural Areas Accidents - Urban Areas are more
```

```python
acc_time=sns.countplot(x= 'Time' ,data=df)
sns.set(rc={'figure.figsize':(50,5)})
for bars in acc_time.containers:
     acc_time.bar_label(bars)
```

```python
Roadtype_speed=df.groupby(['Road_Type'],as_index=False)['Speed_limit'].sum().sort_values(by='Speed_limit',ascending=False).head(10)
sns.set(rc={'figure.figsize':(8,5)})
sns.barplot(data=Roadtype_speed,x='Road_Type',y='Speed_limit')
```

```python
Accident_speed=df.groupby(['Accident_Severity'],as_index=False)['Speed_limit'].sum().sort_values(by='Speed_limit',ascending=False)
sns.barplot(data=Accident_speed,x='Accident_Severity',y='Speed_limit')
```
